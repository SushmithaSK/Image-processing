#Histogram

To plot an image histogram we can write a simple code or use inbuilt functions in cv2 lib.



```python
img = cv2.imread('/content/ghibli.jpg')
```

```python
#plot a histogram
#Syntax: cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
histogram_image = cv2.calcHist([img],[0],None,[256],[0,256])

hist,bins = np.histogram(img.ravel(),256,[0,256])
np.shape(hist)
```

```python
hist[1:10]

#array([ 123,  187,  212,  327,  484,  611,  687,  970, 1617])
#pixel with value 1 -> 123 , 2 -> 187 .....
```

```python
#flatten the histogram
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
#plotting a histogram for each color channel
color =['r','g','b']

for i,col in enumerate(color):
  hist = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist, color=col)
  plt.xlim([0,256])
plt.show()
```
